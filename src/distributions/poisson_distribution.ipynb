# Poisson Distribution


[Wiki](https://en.wikipedia.org/wiki/Poisson_distribution) outlines the following properties.

The probability mass function(PMF) is

$\begin{align*}\quad\quad
f(k) = \frac{\lambda^ke^{-\lambda}}{k!}
\end{align*}$

The cumulative distribution function is

$\begin{align*}\quad\quad
F(k) = \frac{\Gamma(\lfloor k + 1 \rfloor, \lambda)}{\lfloor k \rfloor !}
\end{align*}$

where the upper incomplete gamma function is defined as:

$\begin{align*}\quad\quad
\Gamma(s, x) = \int_x^\infty t^{s - 1} e^{-t} dt
\end{align*}$

## Properties
- Mean is $\lambda$
- Variance is $\lambda$
- Standard Deviation is $\sqrt{\lambda}$
- Skewness is $\lambda^{-1/2}$
- Kurtosis is $3 + \lambda^{-1}$

Moments:
- 1st moment is $\lambda$
- 2nd moment is $\lambda (1 + \lambda)$
- 3rd moment is $\lambda (1 + 3\lambda + \lambda^2)$
- 4th moment is $\lambda (1 + 7\lambda + 6\lambda^2 + \lambda^3)$

If $X_1 \sim Poisson(\lambda_1)$ and $X_2 \sim Poisson(\lambda_2)$ are independent, then the difference $Y = X_1 - X_2$ follows a Skellam distribution.

If $X_1 \sim Poisson(\lambda_1)$ and $X_2 \sim Poisson(\lambda_2)$ are independent, then the distribution of $X_1$ conditional on $X_1 + X_2$ is a binomial distribution.

```python
%reload_ext autoreload
%autoreload 2

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 20)
u = 4
x = np.arange(0, 20)

dist = stats.poisson(u)
plt.title(f'Poisson Distribution', fontsize='16')

plt.plot(x, dist.pmf(x), "ro-")
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 20)
x = np.arange(0, 20)

plt.title(f'Poisson Distribution', fontsize='16')

plt.plot(x, stats.poisson(mu=1).pmf(x), 'ro-', label=r'$\lambda = 1$',)
plt.plot(x, stats.poisson(mu=4).pmf(x), 'bo-', label=r'$\lambda = 4$',)
plt.plot(x, stats.poisson(mu=10).pmf(x), 'go-', label=r'$\lambda = 10$',)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 20)
x = np.arange(0, 20)

plt.title(f'Poisson Distribution', fontsize='16')

plt.plot(x, stats.poisson(mu=1).cdf(x), 'ro-', label=r'$\lambda = 1$',)
plt.plot(x, stats.poisson(mu=4).cdf(x), 'bo-', label=r'$\lambda = 4$',)
plt.plot(x, stats.poisson(mu=10).cdf(x), 'go-', label=r'$\lambda = 10$',)

plt.legend()
plt.show()
```
