# Geometric Distribution



[Wiki](https://en.wikipedia.org/wiki/Geometric_distribution) outlines the following properties.

The probability mass function(PMF) is

$\begin{align*}\quad\quad
f(k) = (1-p)^{k-1}p
\end{align*}$

where $0 < p \leqslant 1$.

The cumulative distribution function is

$\begin{align*}\quad\quad
F(x) = 1-(1-p)^k
\end{align*}$

## Properties
- Mean is $\frag{1}{p}$
- Variance is $\frac{1-p}{p^2}$
- Skewness is $\frac{2-p}{\sqrt{1-p}}$
- Kurtosis is $9 + \frac{p^2}{1-p}$

```python
%reload_ext autoreload
%autoreload 2

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 10)
x = np.arange(1, 10)

dist = stats.geom(p=0.6)
plt.title(f'Geometric Distribution', fontsize='16')

plt.plot(x, dist.pmf(x), "ro-")
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 10)
x = np.arange(1, 10)

plt.title(f'Geometric Distribution', fontsize='16')

plt.plot(x, stats.geom(0.2).pmf(x), 'ro-', label=r'$p=0.2$',)
plt.plot(x, stats.geom(0.5).pmf(x), 'bo-', label=r'$p=0.5$',)
plt.plot(x, stats.geom(0.8).pmf(x), 'go-', label=r'$p=0.8$',)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 10)
x = np.arange(1, 10)

plt.title(f'Geometric Distribution', fontsize='16')

plt.plot(x, stats.geom(0.2).cdf(x), 'ro-', label=r'$p=0.2$',)
plt.plot(x, stats.geom(0.5).cdf(x), 'bo-', label=r'$p=0.5$',)
plt.plot(x, stats.geom(0.8).cdf(x), 'go-', label=r'$p=0.8$',)

plt.legend()
plt.show()
```
