# Hypergeometric Distribution


[Wiki](https://en.wikipedia.org/wiki/Hypergeometric_distribution) outlines the following properties.

The probability mass function(PMF) is

$\begin{align*}\quad\quad
f(k) = \frac{{K \choose k}{N-K \choose n-k}}{N \choose n}
\end{align*}$

where $0 \leqslant K, n \leqslant N$.

The cumulative distribution function(CDF) is

$\begin{align*}\quad\quad
f(k) = 1 - \frac{{n \choose k+1}{N-n \choose K-k-1}}{N \choose K} ~_3F_2(1, k+1-K, k+1-n; k+2, N+k+2-K-n; 1)
\end{align*}$

## Properties
- Mean is $n \frac{K}{N}$
- Variance is $n \frac{K}{N} \frac{N-K}{N} \frac{N-n}{N-1}$
- Skewness is $\frac{(N-2K)(N-1)^{1/2}(N-2n)}{[nK(N-K)(N-n)]^{1/2}(N-2)}$

```python
%reload_ext autoreload
%autoreload 2

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 20)
x = np.arange(0, 20)

dist = stats.hypergeom(500, 50, 100)
plt.title(f'Hypergeometric Distribution', fontsize='16')

plt.plot(x, dist.pmf(x), "ro-")
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 60)
x = np.arange(0, 60)

plt.title(f'Hypergeometric Distribution', fontsize='16')

plt.plot(x, stats.hypergeom(500, 50, 100).pmf(x), 'ro-', label=r'$\lambda = 1$',)
plt.plot(x, stats.hypergeom(500, 60, 200).pmf(x), 'bo-', label=r'$\lambda = 1$',)
plt.plot(x, stats.hypergeom(500, 70, 300).pmf(x), 'go-', label=r'$\lambda = 1$',)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 60)
x = np.arange(0, 60)

plt.title(f'Hypergeometric Distribution', fontsize='16')

plt.plot(x, stats.hypergeom(500, 50, 100).cdf(x), 'ro-', label=r'$\lambda = 1$',)
plt.plot(x, stats.hypergeom(500, 60, 200).cdf(x), 'bo-', label=r'$\lambda = 1$',)
plt.plot(x, stats.hypergeom(500, 70, 300).cdf(x), 'go-', label=r'$\lambda = 1$',)

plt.legend()
plt.show()
```
