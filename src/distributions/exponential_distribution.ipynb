# Exponential Distribution



Let X = amount of time (in minutes) a postal clerk spends with his or her customer.

The value of the change that you have in your pocket or purse approximately follows an exponential distribution.

[Wiki](https://en.wikipedia.org/wiki/Exponential_distribution) outlines the following properties.

The probability dense function(PDF) is

$\begin{align*}\quad\quad
f(x) = \lambda e^{-\lambda x}
\end{align*}$

where $\lambda > 0$ and $x \in [0, \infty)$.

The cumulative distribution function is

$\begin{align*}\quad\quad
F(x) = 1 - e^{-\lambda x}
\end{align*}$

## Properties
- Mean is $\frac{1}{\lambda}$
- Variance is $\frac{1}{\lambda^2}$
- Skewness is $2$
- Kurtosis is $3$

Moments:
$\begin{align*}\quad\quad
E[X^n] = \frac{n!}{\lambda^n}
\end{align*}$

```python
%reload_ext autoreload
%autoreload 2

from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

plt.figure(figsize=(7.5,7.5))
plt.xlabel('Random Variable X', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 5)
x = np.linspace(0, 5, 100)


dist = stats.expon(scale=1)
plt.title(f'Exponential Distribution (Mean = 1)', fontsize='16')

plt.plot(x, dist.pdf(x))
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Random Variable X', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 5)
x = np.linspace(0, 5, 100)

plt.title('Normal Distributions', fontsize='16')

plt.rcParams['text.usetex'] = True
plt.plot(x, stats.expon(scale=1).pdf(x), label=r'$\lambda = 1$')
plt.plot(x, stats.expon(scale=2).pdf(x), label=r'$\lambda = 0.5$')
plt.plot(x, stats.expon(scale=0.5).pdf(x), label=r'$\lambda = 2$')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7.5,7.5))
plt.xlabel('Random Variable X', fontsize='14')
plt.ylabel('Probability', fontsize='14')

plt.xlim(0, 5)
x = np.linspace(0, 5, 100)

plt.title('Normal Distributions', fontsize='16')

plt.rcParams['text.usetex'] = True
plt.plot(x, stats.expon(scale=1).cdf(x), label=r'$\lambda = 1$')
plt.plot(x, stats.expon(scale=2).cdf(x), label=r'$\lambda = 0.5$')
plt.plot(x, stats.expon(scale=0.5).cdf(x), label=r'$\lambda = 2$')
plt.legend()
plt.show()
```
